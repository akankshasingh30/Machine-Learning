```python
#outlier detection

col_list=df.columns
```

```python
for col in col_list:
    if df[col].dtype!='object':    
         sns.boxplot(df[col])
         plt.xlabel(col)
         plt.ylabel("count")
         plt.show()
```

```python
df.columns
```

```python
out_cols=['age', 'duration', 'campaign', 'cons.conf.idx']

for col in out_cols:
    Q1=df[col].quantile(0.25)
    Q3=df[col].quantile(0.75)
    IQR=Q3-Q1
    LL=Q1-1.5*IQR
    UL=Q3+1.5*IQR
    df=df[(df[col]>LL) & (df[col]<UL)]
```

```python
for col in col_list:
    if df[col].dtype!='object':    
         sns.boxplot(df[col])
         plt.xlabel(col)
         plt.ylabel("count")
         plt.show()
```
