```python
# less records so not usual outlier removal instead we perform the exceptional outlier removal
# i.e., also called as percentile method
# instead of taking IQR we will take percentile
```

```python
# in percentile we will compute 1%(0.01) and 99%(0.99) and replace below and above data respectively
```

```python
df.describe()  # see max and min
```

```python
col
```

```python
for i in col:
    Q3= df[i].quantile(0.99)
    Q1= df[i].quantile(0.01)
    df[i][df[i]<=Q1]=Q1
    df[i][df[i]>=Q3]=Q3
```

```python
# now we will perform standardization
# means rescale our data

from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
df_scaled=scaler.fit_transform(new_df)
df_scaled
```

```python
# we got values in array format
# now we will create other df and put the standard values in there
# mean=0
# standard deviation=1


df_df1=pd.DataFrame(df_scaled, columns=['Mat', 'Inns', 'NO', 'Runs', 'HS', 'Ave', 'BF', 'SR', '100', '50', '0','exp'])
df_df1
```

```python
# sum of squared distances or within the cluster sum squared
```

### K MEANS

```python
from sklearn.cluster import KMeans

# elbow curve method to find the ideal no. of clusters
```

```python
clusters=list(range(2,8))
ssd=[]
for num_clusters in clusters:
    model_clus=KMeans(n_clusters=num_clusters, max_iter=150, random_state=50)
    model_clus.fit(df_df1)
    ssd.append(model_clus.inertia_)
```

```python
clusters
```

```python
ssd
```

```python
plt.plot(clusters,ssd, marker='^')
```

```python
# final model

kmean_model=KMeans(n_clusters=4,max_iter=150,random_state=50)
kmean_model.fit(df_df1)
```

```python
kmean_model.labels_
```

```python
df['clusterID']=kmean_model.labels_
df
```

```python
x=df.iloc[:,[3,6]].values
y=kmean_model.fit_predict(x)
```

```python
# visualising the cluster

plt.figure(figsize=(15,7))
sns.scatterplot(x[y==0,0],x[y==0,1],color='yellow',label='cluster1',s=50)
sns.scatterplot(x[y==1,0],x[y==1,1],color='blue',label='cluster1',s=50)
sns.scatterplot(x[y==2,0],x[y==2,1],color='green',label='cluster1',s=50)
sns.scatterplot(x[y==3,0],x[y==3,1],color='grey',label='cluster1',s=50)
plt.xlabel('Not Outs')
plt.ylabel('Average')

sns.scatterplot(kmean_model.cluster_centers_[:4,0],
               kmean_model.cluster_centers_[:4,1],
               color='red',label='centroid',s=300,marker='s')
```
